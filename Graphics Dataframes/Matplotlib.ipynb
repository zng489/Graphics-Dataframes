```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
#step 1 
x = [1,2,3,4]
y = [10,20,25,30]

#step 2
fig = plt.figure() #fig

#step 3
ax = fig.add_subplot(111) #secondary graphics(sub_graphics) ax = fig.add_subplot(size-numeber) #fig

#step 4
ax.plot(x,y, color='red', linewidth=3)
ax.scatter([2,4,6],[5,15,25],color='yellow',marker='^',linewidth=8)

ax.set_xlim(1,6.5)
plt.savefig("foo.png")

#linewidth=(thickness)

plt.title(r"$sigma_i=15$", fontsize=20)

ax.text(1,-2.1, 'example graph', style='italic')

ax.annotate('', xy=(2,10),xycoords='data', 
            xytext=(3,15),textcoords='data', arrowprops=dict(arrowstyle="->", connectionstyle='arc3'),)

#ax.margins(x=2, y=5)
#ax.axis('equal')
#ax.set(xlim=[0,5], ylim=[0,5])
#ax.set_xlim(0,10)

ax.set(title='An example', ylabel='Y-axis', xlabel='X-axis')
ax.legend(frameon=True, title='City',loc='upper left')

plt.savefig("foo.svg", format='svg', dpi=1200)


#step 5
plt.show()
```
